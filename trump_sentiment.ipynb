```python
from transformers import pipeline
# calling and testing the berta
model_name = "cardiffnlp/twitter-roberta-base-sentiment"
sentiment_pipeline = pipeline("sentiment-analysis", model=model_name)
data = ["I love you", "I hate you"]
sentiment_pipeline(data)
```

```python
import pandas as pd
```

```python
trump_dataset = pd.read_csv('cleaned_trump.csv')
```

```python
trump_dataset
```

```python
# narrowing down the biden dataset by deleting all the even rows
trump_dataset = trump_dataset[trump_dataset.index % 2 != 0]
# resetting the index
trump_dataset = trump_dataset.reset_index(drop=True)
trump_dataset
```

```python
# repeat until we get to near 10k rows - second time
trump_dataset = trump_dataset[trump_dataset.index % 2 != 0]
trump_dataset = trump_dataset.reset_index(drop=True)
trump_dataset
```

```python
# repeat again - third time
trump_dataset = trump_dataset[trump_dataset.index % 2 != 0]
trump_dataset = trump_dataset.reset_index(drop=True)
trump_dataset
```

```python
# repeat final time - fourth time
trump_dataset = trump_dataset[trump_dataset.index % 2 != 0]
trump_dataset = trump_dataset.reset_index(drop=True)
trump_dataset
```

```python
# turning tweet column into a list
trump_tweets = trump_dataset.tweet.tolist()
trump_tweets
```

```python
# running a sentiment analysis on the tweets
trump = sentiment_pipeline(trump_tweets)
```

```python
trump = pd.DataFrame(trump)
```

```python
trump
```

```python
trump['index'] = range(1,len(trump)+1)
```

```python
trump_dataset['index'] = range(1, len(trump_dataset)+1)
```

```python
trump_final = pd.merge(trump_dataset, trump, on='index')
```

```python
trump = trump_final
```

```python
trump
```

```python
trump.drop('index', axis=1, inplace=True)
```

```python
trump
```

```python
trump['created_at'] = pd.to_datetime(trump['created_at'])

# Extract just the date part
trump['created_at'] = trump['created_at'].dt.date
trump
```

```python
trump.to_csv('trump_tweets.csv', index=False)
```
