```python
import pandas as pd 
```

```python
# Read dataset
df = pd.read_csv('/Users/olivianguyenvn/Downloads/movies.csv')
```

```python
df
```

```python
# Check dataset information
print(df.head())
```

```python
print(df.info())
```

```python
# Clean dataset by checking if there is any missing value in dataset
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Filling in missing values for categorical columns 
cat_col = df.select_dtypes(include=['object']).columns
for column in cat_col:
    most_frequent_category = df[column].mode()[0]
    df[column].fillna(most_frequent_category, inplace = True)
```

```python
# Filling in missing values for numerical columns 
num_col = df.select_dtypes(include=['int64','float64']).columns
for column in num_col:
    df[column].fillna(df[column].mean(), inplace = True)
```

```python
# Checking missing values again
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Export to new cleaned dataset 
df.to_csv('cleaned_movies.csv', index=False)
```
